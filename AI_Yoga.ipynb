```python
import math
def caluclate_angle(a, b, c):
    a = np.array(a)  # First
    b = np.array(b)  # Mid
    c = np.array(c)  # End

    radians= np.arctan2(c[1]-b[1],  c[0]-b[0])- np.arctan2(a[1]-b[1],  a[0]-b[0])
    angle = np.abs(radians*180/np.pi)

    if(angle>180):
        angle=360-angle
    
    return angle
    
def caluclate_len(a,b):
    a = np.array(a)  # First point
    b = np.array(b)  # Last point

    r_sq = (a[0]-b[0])**2 + (a[1]-b[1])**2+(a[2]-b[2])**2
    r=math.sqrt(r_sq)
    return r
```
